## Generating heatmap for types of problems

```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
problem_levels = [["#1", 5,0,0],["#2",3,2,0],["#3", 0,0,4]]#Difficulty level in three coding competitions -> easy, medium, hard
```

```python
problem_levels
```

```python
df = pd.DataFrame(problem_levels, columns = ["Contest", "Easy", "Medium", "Hard"]).set_index("Contest")
```

```python
df
```

```python
import seaborn as sns
#h_map=sns.set(font_scale=1.5)
h_map=sns.heatmap(df,annot=True,cmap="coolwarm")
h_map.set_xticklabels(h_map.get_xmajorticklabels(), fontsize = 12)
h_map.set_yticklabels(h_map.get_ymajorticklabels(), fontsize = 12,va="center",rotation=0)
h_map.set_ylabel("Coding Contest",fontsize=15)
h_map.set_xlabel("Level of Difficulty",fontsize=15)
h_map.set_title("Varying Levels of Difficulty in Contests",fontsize = 15)
h_map.figure.savefig("plot1.png",dpi=600)
```

### Generating Pie Chart for Number of Registrations

```python
import matplotlib.pyplot as plt
import numpy as np

registered=961
not_registered=1323-961
y = np.array([registered, not_registered])
myexplode = [0.1, 0 ]
patches, texts,_ = plt.pie(y, labels=[f"Registered", f"Not Registered"],autopct='%1.0f%%',explode=myexplode,startangle=45, textprops={'fontsize': 14},colors=["C0","C3"])
plt.legend(patches, [f"Registered {registered}", f"Not Registered {not_registered}"], fontsize=12,loc=3)
plt.title(f"Coding Competition 3 - Registrations",fontsize=18)
plt.savefig("plot2.png",dpi=600)
plt.show()
```

## Generating Pie Chart of Students Who Solved Atleast 1 Problem

```python
import matplotlib.pyplot as plt
import numpy as np

total_registration=961
solved_atleast_one=675
solved_zero_or_not_attempted=total_registration-solved_atleast_one

y = np.array([solved_atleast_one, solved_zero_or_not_attempted])
myexplode = [0.1, 0 ]
patches, texts,_ = plt.pie(y, labels=[f"Solved\nAtleast One", f"Tried But Not\nAble To Solve"],autopct='%1.0f%%',explode=myexplode,startangle=45, textprops={'fontsize': 14},colors=["C0","C3"])
plt.legend(patches, [f"Solved >=1 : {solved_atleast_one}", f"Unable To Solve: {solved_zero_or_not_attempted}"], fontsize=12,loc=3)
plt.title(f"Coding Competition 3 - Solving Challenges",fontsize=18)
plt.savefig("plot3.png",dpi=600)
plt.show()
```

### Registration Into Submission Conversion Rate

```python
registration_to_submission = {"Coding Competition #1": 67.1, "Coding Competition #2" : 66.52, "Coding Competition #3":70.23}
```

```python
import statistics
m=statistics.mean(list(registration_to_submission.values()))
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (9, 7)
plt.bar(list(registration_to_submission.keys()),list(registration_to_submission.values()),color=['darkorange', 'green', 'C0'],width=0.6)#"C0")
l2=plt.axhline(y=m, color='r', linestyle='-')
plot_lines=[]
plot_lines.append([l2])
plt.legend(plot_lines[0],["Average Registration to Submission Percentage"],fontsize=12)
plt.yticks(range(0,101,5))
plt.xticks(fontsize=10)
plt.ylabel("Conversion Percentage",fontsize=16)
plt.xlabel("Various Coding Competitions",fontsize=16)
plt.title("Registration Into Submission Conversion Rate",fontsize=20)
plt.savefig("plot4.png",dpi=600)
```

## Registrations and Submissions Stacked Bar Plot

```python
list_of_competitions=["Coding Competition #1","Coding Competition #2","Coding Competition #3"]
registrations=[1100, 1192, 961]
submissions=[737,793,675]
```

```python
# function to add value labels in bar plot
def addlabels(w,x,y,z):
    for i in range(len(w)):
        plt.text(i,x[i]+y[i],f"Ratio: {z[i]}",ha = 'center',size=14)

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 6)
plt.bar(list_of_competitions,registrations, label="Registrations",width=0.6)#,color=['darkorange', 'green', 'C0'])#"C0")
ax=plt.bar(list_of_competitions,submissions,bottom=registrations, label="Submissions",width=0.6)
addlabels(list_of_competitions,registrations, submissions,[67.1,66.52,70.23])
plt.yticks(range(0,2001,100))
plt.xticks(fontsize=10)
plt.ylabel("Students",fontsize=16)
plt.xlabel("Various Coding Competitions",fontsize=16)
plt.legend(fontsize=12)
plt.title("Stacked Comparison of Registration to Submission",fontsize=18)
plt.savefig("plot5.png",dpi=600)
```

## Branch-wise Summary

```python
import matplotlib.pyplot as plt
# Pie chart
branch = ['CS', 'CSE', 'CE','CSIT', 'ECE','EN', 'IT', 'ME']
count = [ 103, 114, 14+18+5, 101,89+6, 53+18, 95, 31+18+10]

#colors
colors = ['C0','C1','C2','C3', 'C4','C5', 'C6', 'C7']

#explode
explode = (0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05)
 
fig1, ax1 = plt.subplots()
ax1.pie(count, colors = colors, labels=branch, autopct='%1.1f%%', startangle=45,explode=explode,textprops={'fontsize': 14})
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.title(f"Branchwise Summary- Students Who Solved Atleast One Challenge\n",fontsize=18)
plt.savefig("plot7.png",dpi=600,bbox_inches='tight')
plt.show()
```
